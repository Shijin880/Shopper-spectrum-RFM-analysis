```python
from datetime import datetime
import pandas as pd

# Convert InvoiceDate to datetime if it's not already
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Calculate Recency: Days since last purchase
latest_date = pd.Timestamp(datetime.now().date()) # Assuming the analysis is being done now
recency_df = df.groupby('CustomerID')['InvoiceDate'].max().reset_index()
recency_df['Recency'] = (latest_date - recency_df['InvoiceDate']).dt.days

# Calculate Frequency: Number of unique transactions
frequency_df = df.groupby('CustomerID')['InvoiceNo'].nunique().reset_index()
frequency_df.rename(columns={'InvoiceNo': 'Frequency'}, inplace=True)

# Display the first few rows of the recency and frequency dataframes
display(recency_df.head())
display(frequency_df.head())
```

```python
# Merge recency and frequency dataframes
rf_df = recency_df.merge(frequency_df, on='CustomerID')

# Merge the result with the monetary data (which is a Series, so convert to DataFrame first)
rfm_df = rf_df.merge(customer_monetary.reset_index(), on='CustomerID')

# Rename the monetary column for clarity
rfm_df.rename(columns={'Monetary': 'MonetaryValue'}, inplace=True)

# Display the first few rows of the merged dataframe
display(rfm_df.head())
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

wcss=[]
# Drop non-numerical columns before clustering
rfm_clustering = rfm_df.drop(['CustomerID', 'InvoiceDate'], axis=1)
for i in range(1,10):
    Kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42, n_init=10)
    Kmeans.fit(rfm_clustering)
    wcss.append(Kmeans.inertia_)
plt.figure(figsize=(10,10))
plt.plot(range(1,10), wcss,color='red',marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.gca().yaxis.set_major_formatter(ticker.FormatStrFormatter('%.0f'))
plt.show()
```

```python
score = silhouette_score(rfm_clustering, Kmeans.labels_)
```

```python
print('silhouette score:',score)
```
